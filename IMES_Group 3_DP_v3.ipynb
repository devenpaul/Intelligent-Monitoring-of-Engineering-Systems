```python
from kan import KAN
import torch
import copy


seed = 1

model = KAN(width=[6,1,1], grid=3, k=3, seed=seed)

# create dataset


def create_dataset(train_num=500, test_num=500):
    
    def generate_contrastive(x):
        # positive samples
        batch = x.shape[0]
        x[:,2] = torch.exp(torch.sin(torch.pi*x[:,0])+x[:,1]**2)
        x[:,3] = x[:,4]**3

        # negative samples
        def corrupt(tensor):
            y = copy.deepcopy(tensor)
            for i in range(y.shape[1]):
                y[:,i] = y[:,i][torch.randperm(y.shape[0])]
            return y

        x_cor = corrupt(x)
        x = torch.cat([x, x_cor], dim=0)
        y = torch.cat([torch.ones(batch,), torch.zeros(batch,)], dim=0)[:,None]
        return x, y
        
    x = torch.rand(train_num, 6) * 2 - 1
    x_train, y_train = generate_contrastive(x)
    
    x = torch.rand(test_num, 6) * 2 - 1
    x_test, y_test = generate_contrastive(x)
    
    dataset = {}
    dataset['train_input'] = x_train
    dataset['test_input'] = x_test
    dataset['train_label'] = y_train
    dataset['test_label'] = y_test
    return dataset

dataset = create_dataset()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv1D, Dense, Flatten, MaxPooling1D, Dropout, Input
from tensorflow.keras.regularizers import l2
from sklearn.metrics import f1_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt


data = pd.read_csv(r'C:\Users\deven\Desktop\IMES\Project\Merged_df_v4.csv')

data.head(10)
```

```python
# Define input and output variables
X = data[['Time', 'AccX_Filtered','AccY_Filtered','AccZ_Filtered','Section','Frequency','Frequency_Location']]
y = data[['Damping_Location', 'Inclination_Location']]
```

```python
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Reshape input for CNN
X_scaled = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))

# One-hot encode categorical target variables
#y_frequency_loc = to_categorical(y['Frequency_Location'])
y_damping_loc = to_categorical(y['Damping_Location'])
y_inclination_loc = to_categorical(y['Inclination_Location'])

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Separate the outputs for y_train and y_test
#y_train_frequency = y_train['Frequency'].values
#y_train_frequency_loc = to_categorical(y_train['Frequency_Location'].values)
y_train_damping_loc = to_categorical(y_train['Damping_Location'].values)
y_train_inclination_loc = to_categorical(y_train['Inclination_Location'].values)

#y_test_frequency = y_test['Frequency'].values
#y_test_frequency_loc = to_categorical(y_test['Frequency_Location'].values)
y_test_damping_loc = to_categorical(y_test['Damping_Location'].values)
y_test_inclination_loc = to_categorical(y_test['Inclination_Location'].values)
```

```python
from tensorflow.keras.optimizers import Adam

# Define model inputs
inputs = Input(shape=(X_train.shape[1], 1))

# Convolutional base
x = Conv1D(filters=32, kernel_size=2, activation='relu', padding='same')(inputs)
x = MaxPooling1D(pool_size=2, padding='same')(x)
x = Conv1D(filters=64, kernel_size=2, activation='relu', padding='same')(x)
x = MaxPooling1D(pool_size=2, padding='same')(x)
x = Conv1D(filters=128, kernel_size=2, activation='relu', padding='same')(x)
x = MaxPooling1D(pool_size=2, padding='same')(x)
x = Flatten()(x)

# # Dense layers with L2 regularization
# x = Dense(100, activation='relu', kernel_regularizer=l2(0.01))(x)
# x = Dropout(0.25)(x)

# Define multiple outputs
#output_frequency = Dense(1, name='frequency_output')(x)
#output_frequency_loc = Dense(y_train_frequency_loc.shape[1], activation='softmax', name='frequency_loc_output')(x)
output_damping_loc = Dense(y_train_damping_loc.shape[1], activation='softmax', name='damping_loc_output')(x)
output_inclination_loc = Dense(y_train_inclination_loc.shape[1], activation='softmax', name='inclination_loc_output')(x)

# Build the model
model = Model(inputs=inputs, outputs=[  output_damping_loc, output_inclination_loc]) #,output_frequency, output_frequency_loc,

learning_rate = 0.001  # Adjust this value as needed
optimizer = Adam(learning_rate=learning_rate)

# Compile the model
model.compile(optimizer=optimizer, 
              loss={#'frequency_output': 'mse', 
                    #'frequency_loc_output': 'binary_crossentropy',
                    'damping_loc_output': 'binary_crossentropy', 
                    'inclination_loc_output': 'binary_crossentropy'},
              metrics={#'frequency_output': 'mae', 
                       #'frequency_loc_output': 'accuracy', 
                       'damping_loc_output': 'accuracy', 
                       'inclination_loc_output': 'accuracy'})

# Train the model
history = model.fit(X_train, {#'frequency_output': y_train_frequency, 
                              #'frequency_loc_output': y_train_frequency_loc,
                              'damping_loc_output': y_train_damping_loc, 
                              'inclination_loc_output': y_train_inclination_loc},
                    epochs=10, batch_size=64, validation_split=0.2, verbose=1)
```
